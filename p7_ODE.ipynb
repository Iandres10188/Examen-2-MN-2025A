```python
import matplotlib.pyplot as plt
import numpy as np
from src.ODE import ODE_euler  # Asegúrate de que esté bien el path

def f(t, y):
    return -5 * y + 5 * t**2 + 2 * t

def y_exacta(t):
    return t**2 + (1/3) * np.exp(-5 * t)
```

```python
t_vals, y_euler, h = ODE_euler(
    a=0,
    b=1,
    f=f,
    y_t0=1/3,
    N=10
)

y_real = y_exacta(np.array(t_vals))

plt.plot(t_vals, y_euler, 'o--', label='Euler')
plt.plot(t_vals, y_real, '-', label='Exacta')
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Comparación de soluciones (N = 10)")
plt.grid(True)
plt.legend()
plt.show()

print(f"h = {h}")
```

```python
# === CÁLCULO DE ERRORES RELATIVOS PARA N = 5, 10, 20 ===
Ns = [5, 10, 20]
errores = []

for N in Ns:
    t_vals, y_euler, _ = ODE_euler(
        a=0,
        b=1,
        f=f,
        y_t0=1/3,
        N=N
    )

    y_real = y_exacta(np.array(t_vals))
    error_rel = np.mean(np.abs(np.array(y_real) - np.array(y_euler)) / y_real)
    errores.append(error_rel)

for N, e in zip(Ns, errores):
    base, exp = f"{e:.1e}".split("e")
    print(f"e_N={N} = {base} × 10^{int(exp)}")

```
