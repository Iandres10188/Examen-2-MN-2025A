# Splines cúbicos

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 6, "c": -3, "d": 1},     # S0(x): en [-1, 0]
    {"a": 5, "b": -3, "c": 0, "d": -1},    # S1(x): en [0, 1]
]

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 50)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys, color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos (m = -3 en x = 0)")
plt.grid(True)
plt.show()
```
